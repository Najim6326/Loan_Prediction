### 1. Gender (11 missing values):
- **Approach**: Impute with the most frequent value (mode).
- **Reasoning**: Gender is a categorical variable, and imputing with the mode (most common value) is a common approach when the missing data is relatively small. This maintains the distribution of the data without introducing too much bias.

- **Code**:
  ```python
  data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)


### 2. Dependents (10 missing values):
- **Approach**: Impute with the most frequent value (mode).
- **Reasoning**: Like Gender, Dependents is a categorical variable. Imputing with the mode is a simple yet effective method, especially when the number of missing values is low.

- **Code**:
  ```python
  data['Dependents'].fillna(data['Dependents'].mode()[0], inplace=True)


### 3. Self_Employed (23 missing values):
- **Approach**: Impute with the most frequent value (mode).
- **Reasoning**: Similar to Gender and Dependents, Self_Employed is a categorical variable. Since the percentage of missing values is still low, imputing with the mode minimizes bias.

- **Code**:
  ```python
  data['Self_Employed'].fillna(data['Self_Employed'].mode()[0], inplace=True)


### 4. LoanAmount (5 missing values):
- **Approach**: Impute with the median value.
- **Reasoning**: LoanAmount is a numeric variable, and imputing with the median is preferable over the mean because the median is less affected by outliers. This approach preserves the distribution of loan amounts without skewing the data.

- **Code**:
  ```python
  data['LoanAmount'].fillna(data['LoanAmount'].median(), inplace=True)

### 5. Loan_Amount_Term (6 missing values):
- **Approach**: Impute with the most frequent value (mode).
- **Reasoning**: Loan_Amount_Term is a numeric variable but typically takes on discrete values (e.g., 360 months, 180 months). Imputing with the mode (most common loan term) is appropriate since it aligns with the majority of the data.

- **Code**:
  ```python
  data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mode()[0], inplace=True)

### 6. Credit_History (29 missing values):
- **Approach**: Impute with the most frequent value (mode).
- **Reasoning**: Credit_History is a binary variable (1 for good credit, 0 for bad credit). Since credit history is crucial in determining loan eligibility, imputing with the mode (likely 1) minimizes the risk of introducing bias while leveraging the most common outcome.

- **Code**:
  ```python
  data['Credit_History'].fillna(data['Credit_History'].mode()[0], inplace=True)

### Summary of Approaches:

- **Mode Imputation**: Used for categorical variables like `Gender`, `Dependents`, `Self_Employed`, `Loan_Amount_Term`, and `Credit_History` because it preserves the original distribution of the data.

- **Median Imputation**: Used for `LoanAmount` because it reduces the impact of outliers and is more robust than the mean.

These approaches balance the need to retain as much data as possible with the goal of minimizing the introduction of bias. By imputing missing values rather than dropping them, you maintain the dataset's size and integrity, which is especially important if the missing data is relatively small.
    


````python
# Impute missing values for categorical columns using mode
categorical_cols = ['Gender', 'Dependents', 'Self_Employed']
for col in categorical_cols:
    data[col] = data[col].fillna(data[col].mode()[0])

# Impute missing values for numerical columns using mean
numerical_cols = ['LoanAmount', 'Loan_Amount_Term', 'Credit_History']
for col in numerical_cols:
    data[col] = data[col].fillna(data[col].mean())

# Check if all missing values have been handled
missing_values_after_imputation = data.isnull().sum()

# Display the missing values count after imputation
print(missing_values_after_imputation)
````

````python
# Set the style for plots
sns.set(style="whitegrid")

# Univariate Analysis: Histograms for numeric columns
numeric_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']

# Plot histograms
plt.figure(figsize=(15, 5))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, 3, i)
    sns.histplot(data[col], kde=True)
    plt.title(f'Histogram of {col}')

plt.tight_layout()
plt.show()

# Univariate Analysis: Box Plots for numeric columns
plt.figure(figsize=(15, 5))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(data[col])
    plt.title(f'Box Plot of {col}')

plt.tight_layout()
plt.show()

# Univariate Analysis: Bar Charts for categorical columns
categorical_cols = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']

plt.figure(figsize=(18, 10))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(2, 3, i)
    sns.countplot(x=col, data=data)
    plt.title(f'Bar Chart of {col}')

plt.tight_layout()
plt.show()

# Univariate Analysis: Pie Charts for categorical columns
plt.figure(figsize=(18, 10))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(2, 3, i)
    data[col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
    plt.title(f'Pie Chart of {col}')
    plt.ylabel('')

plt.tight_layout()
plt.show()
````
